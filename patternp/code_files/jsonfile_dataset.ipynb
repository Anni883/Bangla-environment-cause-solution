```python
import pandas as pd
import os

# Path to your prepared JSONL
input_file = "/content/drive/MyDrive/softcomproject /DhoroniDataset_for_LLM .jsonl"

# Check the file exists
if not os.path.exists(input_file):
    raise FileNotFoundError(f"File not found at {input_file}")

# Load JSONL into a DataFrame
df = pd.read_json(input_file, lines=True)
print("Rows loaded:", len(df))
print(df.head())
```

Split into train/validation/test


```python
from sklearn.model_selection import train_test_split
import os

# 80% train, 20% temp
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42)

# 10% validation, 10% test
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

print("Train size:", len(train_df))
print("Validation size:", len(val_df))
print("Test size:", len(test_df))

# Save splits to Drive as JSONL
output_dir = "/content/drive/MyDrive/softcomproject /DhoroniFiles" # Corrected path

# Create the directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

train_df.to_json(os.path.join(output_dir, "train.jsonl"), orient="records", lines=True, force_ascii=False)
val_df.to_json(os.path.join(output_dir, "val.jsonl"), orient="records", lines=True, force_ascii=False)
test_df.to_json(os.path.join(output_dir, "test.jsonl"), orient="records", lines=True, force_ascii=False)

print("Saved train/val/test JSONL files in:", output_dir)
```
